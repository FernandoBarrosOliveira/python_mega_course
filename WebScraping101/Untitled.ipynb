```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
req=requests.get("http://versoes.milenioerp.com.br/")
```

```python
req.content
```

```python
soup=BeautifulSoup(req.content, "html.parser")
```

```python
print(soup.prettify)
```

```python
all=soup.find_all("span", {"class":"secondary label"})
all
```

```python
for item in all:
    print(item.text)
```
